```python
# Import pandas
import pandas as pd 
```

```python
# Read in the data file: df
df1 = pd.read_csv('global.csv') #global temperature data
df2=pd.read_csv('city.csv') #Tripoli temperature data

#make it easy to call the data
global_data=df1['avg_temp']
city=df2['avg_temp'] 
```

```python
#view global temperature output 
global_data.head()
```

```python
#view Tripoli temperature output 
city.head()
```

```python
global_mavg=global_data.rolling(10).mean()
print(global_mavg)
```

```python
city_mavg=city.rolling(10).mean()
print(city_mavg)
```

# Plotting the Data 

```python
#import libraries 

import matplotlib.pyplot as plt
from matplotlib import style
```

```python
#Plotting the graph 


plt.xlabel("Date")
plt.ylabel("Avg Temperature") 
plt.title("Trends of Global and Local (Tripoli) Avg Temperatures")

plt.plot(df1['year'],global_mavg,label='Global avg temperatures')
plt.plot(df2['year'],city_mavg,label='Tripoli avg temperatures')

plt.legend()
```

```python
plt.xlabel("Date")
plt.ylabel("Avg Temperature") 
plt.plot(df1['year'],global_mavg,label='Global avg temperatures', color='blue')
plt.title("Global Avg Temperatures")

plt.legend()
```

```python
plt.xlabel("Date")
plt.ylabel("Avg Temperature") 
plt.plot(df2['year'],city_mavg,label='Tripoli avg temperatures')
plt.title("Tripoli Avg Temperatures")
plt.legend()
```
